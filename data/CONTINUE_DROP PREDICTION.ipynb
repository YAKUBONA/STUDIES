```python
from sklearn import preprocessing
```

```python
def categorical_transform(train_data,test_data):
    le = preprocessing.LabelEncoder()
    train_data.gender = le.fit_transform(train_data.gender)
    train_data.student_id = le.fit_transform(train_data.student_id)
    train_data.guardian = le.fit_transform(train_data.guardian)
    train_data.caste = le.fit_transform(train_data.caste)
    test_data.gender = le.fit_transform(test_data.gender)
    test_data.student_id = le.fit_transform(test_data.student_id)
    test_data.guardian = le.fit_transform(test_data.guardian)
    test_data.caste = le.fit_transform(test_data.caste)
    return train_data 
```

```python
X=train_data[[ 'student_id', 'gender', 'caste', 'mathematics_marks',
       'english_marks', 'science_marks', 'science_teacher',
       'languages_teacher', 'guardian', 'internet', 'school_id',
       'total_students', 'total_toilets', 'establishment_year']]
y=train_data['continue_drop']
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
test_data[:5]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42)
```

```python
mlp = MLPClassifier(solver='lbfgs',random_state=0).fit(X_train, y_train)
```

```python
y_pred = mlp.predict(X_val)
```

```python
print("Training set score: {:.7f}".format(f1_score(y_val, y_pred)))
```
